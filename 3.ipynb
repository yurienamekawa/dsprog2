# 被安打被本塁打

```python
import pandas as pd
# データを読み込んで df を定義する
df = pd.read_csv('lecture-sp/stats_pit.csv', encoding='cp932')
# チームごとに合計（hのみ）
team_h = df.groupby('team_name')['h'].sum().reset_index()
print('-- 被安打（h）チームごと --')
print(team_h)
# チームごとに合計（hrのみ）
team_hr = df.groupby('team_name')['hr'].sum().reset_index()
print('-- 被本塁打（hr）チームごと --')
print(team_hr)







```
