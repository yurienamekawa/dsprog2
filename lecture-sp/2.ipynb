# 勝率

```python
import pandas as pd
# 'lecture-sp/' を追加して、フォルダの中を指定します
df = pd.read_csv('lecture-sp/games.csv', encoding='cp932')
# ホーム・アウェイの勝敗判定
df['home_win'] = df['home_team_score'] > df['away_team_score']
df['away_win'] = df['home_team_score'] < df['away_team_score']
df['draw'] = df['home_team_score'] == df['away_team_score']
# ホーム側の戦績
home_result = df[['home_team_name', 'home_win', 'away_win', 'draw']].copy()
home_result.columns = ['team', 'win', 'lose', 'draw']
# アウェイ側の戦績
away_result = df[['away_team_name', 'away_win', 'home_win', 'draw']].copy()
away_result.columns = ['team', 'win', 'lose', 'draw']
# 縦につなげる
all_results = pd.concat([home_result, away_result])
# 各チームごとに合計
team_stats = all_results.groupby('team').sum()
# 勝率計算（引き分けを除く）
team_stats['win_rate'] = team_stats['win'] / (team_stats['win'] + team_stats['lose'])
team_stats['win_rate'] = team_stats['win_rate'].round(3)
print(team_stats[['win', 'lose', 'draw', 'win_rate']])
```
