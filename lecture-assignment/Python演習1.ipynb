### 問3

「ヘルシンキ」＋「エストニア」の文字を先頭から交互に連結して文字列「ヘエルスシトンニキア」を出力してください

```python
b = "ヘルシンキ"
c = "エストニア"
interleaved_chars = [f"{c1}{c2}" for c1, c2 in zip(b, c)]
result_string = "".join(interleaved_chars)
print(result_string)
```

### 問4

"Character education based on the fundamental spirit of Buddhism and the Shiguzeigan"という文を単語に分解し，各単語の（アルファベットの）文字数を先頭から出現順に並べたリストを作成してください


```python
import re
text = "Character education based on the fundamental spirit of Buddhism and the Shiguzeigan"
# 句読点などを削除して単語をスペースで区切る
cleaned_text = re.sub(r'[^\w\s]', '', text)
# スペースで単語に分割する
words = cleaned_text.split()
# 各単語の文字数を取得して、リストにする
word_lengths = [len(word) for word in words]
print(f"単語の分解: {words}")
print(f"文字数リスト: {word_lengths}")
```

### 問5

"It is not the strongest of the species that survives, nor the most intelligent that survives. It is the one that is most adaptable to change."という文を単語に分解し，1, 2, 5, 6, 8, 9, 16番目の単語の先頭の2文字を取り出し，取り出した文字列から単語の位置（先頭から何番目の単語か）を格納した辞書を作成してください

```python
original_text = "It is not the strongest of the species that survives, nor the most intelligent that survives. It is the one that is most adaptable to change."
# コンマとピリオドの除去
text_without_punctuation = original_text.replace(',', '').replace('.', '')
# 単語への分解
word_list = text_without_punctuation.split()
target_positions_1_based = [1, 2, 5, 6, 8, 9, 16]
# 0から始まるインデックスへ変換
target_indices_0_based = [position - 1 for position in target_positions_1_based]
# 指定位置の単語から先頭2文字を抽出
extracted_two_chars = [word_list[index][:2] for index in target_indices_0_based]
# 辞書の作成
word_position_map = dict(zip(extracted_two_chars, target_positions_1_based))
print(word_position_map)
```
