# 投手打者

```python
import pandas as pd
df = pd.read_csv('lecture-sp/players.csv', encoding='cp932')
# 投手分布（投げる手）
pitchers = df[df['position_type'] == '投手']
pitcher_hand_dist = pitchers.groupby(['team_name', 'throws']).size().unstack(fill_value=0)
print("投手の左右分布（各チーム）\n", pitcher_hand_dist)
# 打者分布（打つ手）
batters = df[df['position_type'] != '投手']
batter_hand_dist = batters.groupby(['team_name', 'bats']).size().unstack(fill_value=0)
print("打者の左右分布（各チーム）\n", batter_hand_dist)








```
